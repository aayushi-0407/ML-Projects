####import all necessary modules

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.decomposition import PCA
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans , AgglomerativeClustering

```

```python
data=pd.read_csv("/content/wine-clustering.csv")
```

```python
data.describe()
```

####NO null values are present

```python
data.info()
```

####Box plot

```python
for col in data.columns:
  plt.figure(figsize=(6,6))
  plt.title(col)
  plt.boxplot(data[col])
  plt.show()

```

####No as such outliers are present

####Correlation Matrix using heat map

```python
plt.figure(figsize=(10,12))
corr=data.corr()
sns.heatmap(corr,annot=True,vmin=-1,vmax=1,center=0)
```

```python
data_copy1=data.copy() #creating copy of data
```

####principal component analysis

```python
pca=PCA(n_components=2)
pca_data=pca.fit_transform(data)
pca_df=pd.DataFrame(data=pca_data,columns=['pca1','pca2'])
```

####plottinng dataset

```python
plt.figure(figsize=(8,6))
plt.scatter(pca_df.iloc[:,0],pca_df.iloc[:,1])
plt.title("2D Plot of PCA data")

```

####principal component analysis using 3 components

```python
pca=PCA(n_components=3)
pca_data=pca.fit_transform(data)
pca_df=pd.DataFrame(data=pca_data,columns=['pca1','pca2','pca3'])
```

####plotting graph

```python
fig=plt.figure(figsize=(10,6))
ax=fig.add_subplot(111,projection='3d')
ax.scatter(pca_df['pca1'],pca_df['pca2'],pca_df['pca3'])
ax.set_title("Projection of data in 3D")
plt.show()
```

####using elbow method to visulaise number of clusters(K)

```python
elbow=KElbowVisualizer(KMeans(),k=10)
elbow.fit(pca_df)
elbow.show()
```

####Model 1 Agglomerative Clustering

```python
ac1=AgglomerativeClustering(n_clusters=4)
ac1_predict=ac1.fit_predict(pca_df)
ac1_df=pca_df.copy()
ac1_df['cluster']=ac1_predict
```

```python
ac1_df
```

```python
fig=plt.figure(figsize=(10,6))
ax=fig.add_subplot(111,projection='3d')
ax.scatter(ac1_df['pca1'],ac1_df['pca2'],ac1_df['pca3'],c=ac1_df['cluster'],cmap='Accent')
ax.set_title("Projection of data in 3D")
plt.show()
```

####Distribution of clusters

```python
fig=sns.countplot(x=ac1_df['cluster'],palette='Accent')
fig.set_title("Distribution of cluster in ac1")
```

####Model 2 Kmeans

```python
kmeans1=KMeans(n_clusters=4)
kmeans1_pred=kmeans1.fit_predict(pca_df)
kmeans1_df=pca_df.copy()
kmeans1_df['cluster']=kmeans1_pred
```

```python
fig=plt.figure(figsize=(10,6))
ax=fig.add_subplot(111,projection='3d')
ax.scatter(kmeans1_df['pca1'],kmeans1_df['pca2'],kmeans1_df['pca3'],c=kmeans1_df['cluster'],cmap='Accent')
ax.set_title("Projection of data in 3D")
plt.show()
```

####Cluster distribution

```python
fig=sns.countplot(x=kmeans1_df['cluster'],palette='Accent')
fig.set_title("Distribution of cluster in kmeans1")
```
