```python
import pandas as pd
```

```python
df=pd.read_csv('/content/ecommerceDataset.csv', names=["Category","Description"], header=None)
```

```python
df.head()
```

```python
df.Category.value_counts()
```

```python
df.dropna(inplace=True)
df.shape
```

```python
df.Category.replace("Clothing & Accessories", "Clothing_Accessories", inplace=True)
```

```python
df['Category']="__label__"+df['Category'].astype(str)
df.head(3)
```

```python
df['category_description']=df['Category']+" "+df['Description']
df.head(3)
```

```python
text=df.category_description[0]
```

```python
import re
def preprocess(text):
  text=re.sub(r'[^\w\s\']',' ',text)
  text=re.sub(r' +',' ',text)
  return text.strip().lower()
```

```python
preprocess(text)
```

```python
df['category_description']=df['category_description'].apply(preprocess)
```

```python
df.head(4)
```

```python
from sklearn.model_selection import train_test_split
train,test=train_test_split(df,test_size=0.2)
```

```python
train.shape,test.shape
```

```python
train.to_csv('train.csv',header=False,index=False, columns=['category_description'])
test.to_csv('test.csv',header=False,index=False, columns=['category_description'])
```
